````python
# date_cols = ["order_date_time", "Issue_reported at", "issue_responded", "Survey_response_Date"]

# for col in date_cols:
#     if col in df.columns:
#         df[col] = pd.to_datetime(df[col], errors="coerce")


import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Load the data
df = pd.read_csv(file_path)

# Basic exploration
print(df.info())
print(df.describe())
print(df.isnull().sum())

# Handle missing values
# For numerical columns
df_numeric = df.select_dtypes(include=['number'])
numeric_cols = df_numeric.columns.tolist()
for col in numeric_cols:
    # Replace missing values with median
    df[col].fillna(df[col].median(), inplace=True)
    
# For categorical columns
df_categorical = df.select_dtypes(include=['object'])
categorical_cols = df_categorical.columns.tolist()
for col in categorical_cols:
    # Replace missing values with most frequent value
    df[col].fillna(df[col].mode()[0], inplace=True)

# Convert date columns to datetime
date_cols = ["date_column1", "date_column2"]  # Replace with your date column names
for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors="coerce")

# Handle outliers using IQR method for numeric columns
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[col] = np.where((df[col] < lower_bound) | (df[col] > upper_bound),
                       df[col].median(),
                       df[col])

# Create a preprocessing pipeline
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])

# Apply preprocessing
X = df.drop('target_column', axis=1)  # Replace 'target_column' with your target
y = df['target_column']  # Replace with your target column
X_processed = preprocessor.fit_transform(X)

print("Preprocessing complete!")
````

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Hypothetical Statement - 1 : CSAT Scores remain constant across Agent Shifts

Null Hypothesis (H₀): The average CSAT score is the same across all agent shifts (e.g., Morning, Afternoon, Night).

Alternate Hypothesis (H₁): At least one agent shift has a different average CSAT score.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

# Step 1: Clean column names (remove hidden spaces)
df.columns = df.columns.str.strip()

# Step 2: Combine all Agent Shift columns into one column
shift_cols = [c for c in df.columns if "Agent Shift_" in c]

df["agent_shift"] = df[shift_cols].idxmax(axis=1)
df["agent_shift"] = df["agent_shift"].str.replace("Agent Shift_", "", regex=False)

# Step 3: Rename CSAT column
df = df.rename(columns={"CSAT Score": "csat_score"})

# Step 4: Run ANOVA
from statsmodels.formula.api import ols
import statsmodels.api as sm

model = ols("csat_score ~ C(agent_shift)", data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print(anova_table)
````

##### Which statistical test have you done to obtain P-Value?

ANOVA is used to compare the means of a numeric variable (CSAT) across more than two independent groups (shifts). It's ideal when the independent variable is categorical and the dependent variable is continuous.

##### Why did you choose the specific statistical test?

There is strong evidence that agent shift has a significant effect on CSAT Score since the p-value is extremely small (<< 0.05), and the F-statistic is large. This means CSAT Scores differ across different agent shifts in a way unlikely due to random chance.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H₀): Product category and issue category are independent (no association).

Alternate Hypothesis (H₁): Product category and issue category are associated.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

from scipy.stats import chi2_contingency
import pandas as pd

# Clean column names
df.columns = df.columns.str.strip()

# Combine one-hot encoded product category columns
prod_cols = [c for c in df.columns if "Product_category_" in c]

df["product_category"] = df[prod_cols].idxmax(axis=1)
df["product_category"] = df["product_category"].str.replace("Product_category_", "", regex=False)

# Rename CSAT column
df.rename(columns={"CSAT Score": "csat_score"}, inplace=True)

# Create contingency table
contingency_table = pd.crosstab(df["product_category"], df["csat_score"])

# Chi-Square Test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-square statistic:", chi2)
print("Degrees of freedom:", dof)
print("P-value:", p)
````

##### Which statistical test have you done to obtain P-Value?

Both product category and issue type are categorical. The Chi-square test is appropriate to assess whether two categorical variables are related.

````python
# Handling Outliers & Outlier treatments
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Numeric columns in your dataset
cols = ['Item_price', 'connected_handling_time', 'response_time_hours', 'csat_score']

# Detect and treat outliers using IQR
for col in cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    # Capping outliers
    df[col] = np.where(df[col] < lower, lower, df[col])
    df[col] = np.where(df[col] > upper, upper, df[col])

# Visualization after treatment
plt.figure(figsize=(10,6))
sns.boxplot(data=df[['Item_price','connected_handling_time','response_time_hours','csat_score']])
plt.title("Boxplot After Outlier Treatment")
plt.show()
````
